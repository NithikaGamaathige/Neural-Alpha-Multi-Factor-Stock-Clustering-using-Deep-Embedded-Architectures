### Project Initialization & Universe Selection

```python
# --- CELL 1: SETUP ---
import yfinance as yf
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from datetime import datetime, timedelta
import matplotlib.pyplot as plt # Kept for the static plots if needed

# [cite_start]1. Define Tickers (Group 2) [cite: 6]
tickers = [
    "CMCSA", "GOOG", "NFLX", "T", "VZ", "AZO", "CVNA", "DRI", "GPC", "LEN",
    "MCD", "ORLY", "ROST", "TSLA", "ADM", "CL", "DLTR", "KDP", "KVUE", "MO",
    "STZ", "WMT", "CVX", "FANG", "OXY", "VLO", "AIG", "AON", "BEN", "BX",
    "CINF", "EG", "FITB", "HIG", "JKHY", "MA", "MSCI", "PGR", "RJF", "TFC",
    "WFC", "ABBV", "BDX", "CI", "CVS", "ELV", "HOLX", "IQV", "MCK", "MTD",
    "RVTY", "TMO", "WAT", "ALLE", "BLDR", "CMI", "HUBS", "EMR", "FIX", "GNRC",
    "HWM", "JBHT", "LMT", "NOC", "PAYX", "ROK", "SWK", "UBER", "VRSK", "ACN",
    "AMAT", "AVGO", "CSCO", "FFIV", "GEN", "INTC", "KLAC", "MSI", "NXPI", "PTC",
    "TEL", "VRSN", "AMCR", "CRH", "FCX", "MLM", "PPG", "AMT", "CCI",
    "EQIX", "HST", "O", "SPG", "WY", "AWK", "DTE", "ETR", "NEE", "PNW", "WEC"
]

# Technical Guardrail: Alphabetical Sorting & Uniqueness
tickers = sorted(list(set(tickers)))
benchmark_ticker = "^GSPC" # S&P 500

# 2. Define Timeframe (Explicit 2-Year Window)
end_date = datetime(2026, 2, 8)
start_date = end_date - timedelta(days=365*2)

print(f"Setup Complete. {len(tickers)} tickers prepared.")
```

### Data Acquisition & Preprocessing:

```python
# --- CELL 2: DATA INGESTION ---
# This cell relies on 'tickers', 'benchmark_ticker', 'start_date', and 'end_date' 
# from Cell 1. Make sure you have run Cell 1 successfully first.

print(f"Downloading data from {start_date.date()} to {end_date.date()}...")

# Download everything at once to save time
# We combine the stock tickers and the benchmark into one request
all_data = yf.download(tickers + [benchmark_ticker], start=start_date, end=end_date, auto_adjust=True, progress=True)['Close']

# Separate Benchmark from Stocks
df_benchmark = all_data[benchmark_ticker]
df_stocks = all_data[tickers]

# Technical Guardrail: Remove columns that are 100% empty (e.g., delisted stocks)
df_clean = df_stocks.dropna(axis=1, how='all')

# Fill small gaps (weekend/holiday artifacts)
df_clean = df_clean.ffill().bfill()
df_benchmark = df_benchmark.ffill().bfill()

print(f"Data Download Complete. Active Stocks: {len(df_clean.columns)}")
print(f"Shape: {df_clean.shape}")
```

### EXPORT CLEANED DATASETS

```python
# --- EXPORT FOR NOTEBOOK 2 ---
# We save both stocks and benchmark to separate files
df_clean.to_parquet('group2_stocks.parquet')
df_benchmark.to_frame().to_parquet('benchmark.parquet')

print("✅ Data persisted to local directory. Ready for Notebook 2.")
```
